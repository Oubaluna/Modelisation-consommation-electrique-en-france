# Gestion et check des types de données

```python
df['Date'] = pd.to_datetime(df['Date'], utc=True)

df.info
```

# Gestion des Nans & Uniformisation Colonnes

Remplacement par la médiane pour éviter d'erroné la distribution des variables. 

```python
l_col_nan = (df.isna().sum()/df.shape[0]).sort_values(ascending=False).index[0:10]

for i in l_col_nan:
    df[i] = df[i].fillna(df[i].median())

(df.isna().sum()/df.shape[0]).sort_values(ascending=False)
```

```python
liste_normalize = utils.normalize(df.columns)
liste_trad = utils.list_to_dict(liste_normalize)

df = df.set_axis([i for i in range(0,17)], axis=1, inplace=False)
df = df.rename(columns=liste_trad)

df.columns
```

# Concaténation des données de prod_elec et de Météo

### Grouper les données productions élec par jour par région

```python
prod_elec['date'] =  pd.to_datetime(prod_elec['date'])
prod_elec['year'] = prod_elec['date'].apply(lambda x : x.year)
prod_elec['month'] = prod_elec['date'].apply(lambda x : x.month)
prod_elec['days'] = prod_elec['date'].apply(lambda x : x.day)

prod_elec_jour = prod_elec.groupby(['year', 'month', 'days', 'region'], as_index = False).sum()
prod_elec_jour = prod_elec_jour.drop('code_insee_region', axis=1)

prod_elec_jour_v2 = prod_elec.groupby(['year', 'month', 'days'], as_index = False).sum()
prod_elec_jour_v2 = prod_elec_jour_v2.drop('code_insee_region', axis=1)
prod_elec_jour_v2
```

```python
prod_elec_jour['date'] = prod_elec_jour['year'].astype(str) + '-' + prod_elec_jour['month'].astype(str) + '-' + prod_elec_jour['days'].astype(str)
prod_elec_jour['date'] =  pd.to_datetime(prod_elec_jour['date'])
prod_elec_jour = prod_elec_jour.drop(['year', 'month', 'days'], axis=1)

prod_elec_jour_v2['date'] = prod_elec_jour_v2['year'].astype(str) + '-' + prod_elec_jour_v2['month'].astype(str) + '-' + prod_elec_jour_v2['days'].astype(str)
prod_elec_jour_v2['date'] =  pd.to_datetime(prod_elec_jour_v2['date'])
prod_elec_jour_v2 = prod_elec_jour_v2.drop(['year', 'month', 'days'], axis=1)
```

## Grouper les données de météo par jour par region

Si besoin : 
0° = nord (vent du nord) [5]

45° = nord-est (vent du nord-est) [6]

90° = est (vent d'est) [7]

135° = sud-est (vent du sud-est) [8]

180° = sud (vent du sud) [1]

225° = sud-ouest (vent du sud-ouest) [2]

270° = ouest (vent d'ouest) [3]

315° = nord-ouest (vent du nord-ouest) [4]

360° = nord [5]

```python
df['year'] = df['date'].apply(lambda x : x.year)
df['month'] = df['date'].apply(lambda x : x.month)
df['days'] = df['date'].apply(lambda x : x.day)

df['year'].unique()
```

```python
function_to_apply = {
    'rafales_sur_une_periode' : 'max', 
    'variation_de_pression_en_3_heures' : 'sum',
    'precipitations_dans_les_3_dernieres_heures' : 'sum', 
    'regioncode' : 'sum', # à jeter
    'pression_station' : 'mean', 
    'pression_au_niveau_mer' : 'mean', 
    'direction_du_vent_moyen' : 'median', #médiane de la direction sur la journée
    'vitesse_du_vent_moyen' : 'mean', 
    'humidite' : 'mean', 
    'point_de_rosee' : 'mean', 
    'temperature_c' : 'mean',
    'latitude' : 'median', 
    'longitude' : 'median', 
    'altitude' : 'median', 
}

meteo_jour = df.groupby(['year', 'month', 'days', 'region'], as_index = False).agg(function_to_apply)

meteo_jour_v2 = df.groupby(['year', 'month', 'days'], as_index = False).agg(function_to_apply)

df3h = df.groupby('date', as_index=False).agg(function_to_apply)

## EXPORT
df3h.to_csv('C:\\Users\\Nouz\\Desktop\\DATASCIENTEST\\Projet Energie\\clean_data\\clean_data_meteo_3h.csv',index=False)
```

```python
meteo_jour['date'] = meteo_jour['year'].astype(str) + '-' + meteo_jour['month'].astype(str) + '-' + meteo_jour['days'].astype(str)
meteo_jour['date'] =  pd.to_datetime(meteo_jour['date'])
meteo_jour = meteo_jour.drop(['year', 'month', 'days'], axis=1)

meteo_jour_v2['date'] = meteo_jour_v2['year'].astype(str) + '-' + meteo_jour_v2['month'].astype(str) + '-' + meteo_jour_v2['days'].astype(str)
meteo_jour_v2['date'] =  pd.to_datetime(meteo_jour_v2['date'])
meteo_jour_v2 = meteo_jour_v2.drop(['year', 'month', 'days'], axis=1)
```

# MERGE avec la méthode INNER pour obtenir un dataset clean de NaN en utilisant la journée et la région

```python
data_meteo_prod = pd.merge(left=meteo_jour, right=prod_elec_jour, right_on=['date', 'region'], left_on=['date', 'region'], how='inner')

data_meteo_prod_v2 = pd.merge(left=meteo_jour_v2, right=prod_elec_jour_v2, right_on=['date'], left_on=['date'], how='inner')

data_meteo_prod_v2
```
